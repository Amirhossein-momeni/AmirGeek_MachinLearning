```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
######################################################################################
# Reading Data
######################################################################################

# Reading the CSV file into a DataFrame
df = pd.read_csv("../0000_Data/01-raw/01-Ad.csv")

# Defining feature columns and target variable
x = df.iloc[:, 3:5].values  
y = df.iloc[:,-1].values 
```

```python
######################################################################################
# Preprocessing
######################################################################################

# Scaling features using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
######################################################################################
# Building the Model with KNeighborsClassifier(KNN)
######################################################################################

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x, y)

# ==============================================================================
# Concept: KNN (K-Nearest Neighbors) - Classification
# ==============================================================================
# - Goal: Classifies a new data point based on the majority vote of its
#   'k' closest neighbors in the feature space.
#
# - n_neighbors (k): The number of neighbors to consider (e.g., 5).
#   Small k can be noisy; large k can be too general.
#
# - Fitting: The model calculates distances between all training points.
#   CRITICAL: Input features must be SCALED so that distance calculations
#   are fair and not dominated by large-range features.
# ==============================================================================
```
