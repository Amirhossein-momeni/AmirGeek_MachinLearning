```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#####################################################################
# Reading Data
#####################################################################
# Reading the CSV file into a DataFrame
df = pd.read_csv("../0000_Data/01-raw/01-Ad.csv")
# Defining feature columns and target variable
x = df.iloc[:, 3:5].values  
y = df.iloc[:,-1].values 
```

```python
#####################################################################
# Preprocessing
#####################################################################
# Scaling features using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
#####################################################################
# Building the Model with KNeighborsClassifier(KNN)
#####################################################################
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x, y)
# KNN : K-Nearest Neighbors algorithm for classification tasks.
# It classifies a data point based on the majority class among its k-nearest neighbors in the feature space.
# n_neighbors=5 specifies that the 5 nearest neighbors will be considered for making predictions.
# Fitting the model : The model is (trained) on __"the scaled feature set x"__ and __"target variable y"__.
```
