```python
# min max scaling:

# x = (x - min(x)) / (max(x) - min(x))
# Min-Max scaler scales the data between 0 and 1.
# For Min-Max scaling, we (must) have a (two)-dimensional array.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x = np.array([
    [1, 10000000],
    [2, 20000000],
    [3, 30000000],
    [4, 40000000]
])

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
# Fit means find the min and max values and we have: fit(x) method
# Transform means scale the data between 0 and 1 and we have: transform(x) method
```
