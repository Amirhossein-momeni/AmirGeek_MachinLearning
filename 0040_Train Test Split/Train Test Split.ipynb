```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
###########################################################################
# Reading Data
###########################################################################
# Reading the CSV file into a DataFrame
df = pd.read_csv("../0000_Data/01-raw/01-Ad.csv")

# Defining feature columns and target column
x_values = df.iloc[:, 3:5].values
y_values = df.iloc[:, -1].values
```

```python
# What are X_values and y_values?
# X_values are the feature columns, and y_values are the target variable.
# X_values are the columns that we will use to predict the target variable, which is y_values.
# y_values is the column that contains the target variable that we want to predict.
```

```python
###########################################################################
# Preprocessing
###########################################################################
# Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.25, stratify=y_values, random_state=42)

# What is the training process in ML? : Fitting the model on training data so that it can learn patterns. 
# What is testing process in ML? : Evaluating the model on unseen data to check its performance.

# What is x_train? : x_train is the portion of the feature data that is used to train the machine learning model. It contains the input features that the model will learn from during the training process.
# What is y_train? : y_train is the portion of the target variable data that corresponds to x_train. It contains the actual labels or values that the model will learn to predict during training.
# What is x_test? : x_test is the portion of the feature data that is used to evaluate the performance of the trained machine learning model. It contains input features that the model has not seen during training, allowing us to assess how well the model generalizes to new data.
# What is y_test? : y_test is the portion of the target variable data that corresponds to x_test. It contains the actual labels or values for the test set, which are used to evaluate the model's predictions and measure its performance.

# What is test Size? : 25% of the data is used for testing, meaning 75% is used for training.
# What is Stratify? : Stratify is a parameter in train_test_split that ensures the class distribution in the target variable is maintained in both training and testing sets, which is crucial for imbalanced datasets. It helps to ensure that the proportion of classes in the target variable is consistent across both sets, preventing bias in model evaluation.
# So stratify keeps the same class percentages in train and test data.

# Sclaling Features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Why we don't need to fit on x_test?
# We fit preprocessing steps only on the training data to learn the required parameters (e.g., mean, variance).
# The test data must not be used for fitting because it represents unseen data; otherwise, data leakage occurs.
# Therefore, we only transform X_test using the parameters learned from X_train.
```

```python
###########################################################################
# Model Training
###########################################################################
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=7)
KNN.fit(x_train, y_train)
```
