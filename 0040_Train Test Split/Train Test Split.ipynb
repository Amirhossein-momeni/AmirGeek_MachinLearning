```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
###########################################################################
# Reading Data
###########################################################################
# Reading the CSV file into a DataFrame
df = pd.read_csv("../0000_Data/01-raw/01-Ad.csv")

# Defining feature columns and target column
x_values = df.iloc[:, 3:5].values
y_values = df.iloc[:, -1].values
```

```python
###########################################################################
# Preprocessing
###########################################################################
# Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.25, stratify=y_values, random_state=42)
# What is the training in ML? : Fitting the model on training data so that it can learn patterns. 
# What is testing in ML? : Evaluating the model on unseen data to check its performance.
# Why we nees to x_train, x_test, y_train, y_test :: x_train and y_train are used to train the model, while x_test and y_test are used to evaluate its performance on unseen data.
# Test Size : 25% of the data is used for testing, meaning 75% is used for training.
# What is Stratify? : Stratify is a parameter in train_test_split that ensures the class distribution in the target variable is maintained in both training and testing sets, which is crucial for imbalanced datasets. It helps to ensure that the proportion of classes in the target variable is consistent across both sets, preventing bias in model evaluation.
# Stratify keeps the same class percentages in train and test data.

# Sclaling Features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
# Why we don't need to fit on x_test? : Because we want to apply the same scaling parameters (mean and standard deviation) learned from the training data to the test data, ensuring consistency in feature scaling.
```

```python
###########################################################################
# Model Training
###########################################################################
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=7)
KNN.fit(x_train, y_train)
```
