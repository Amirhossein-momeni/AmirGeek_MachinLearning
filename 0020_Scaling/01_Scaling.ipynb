```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# why do we need scaling?
# to make the data in the same range
# to make the data in the same distribution
# scaling means removing the effect of measurements units from the data
```

```python
# Types of scaling :

# min max scaling
# standard scaling
# z-score scaling
```

```python
# ===================================================================================
# min max scaling:
# ===================================================================================

# x = (x - min(x)) / (max(x) - min(x))
# Min-Max scaler scales the data between 0 and 1.
# For Min-Max scaling, we (must) have a (two)-dimensional array.
# If you want to scale a one-dimensional array, you need to reshape it first.
# reshape(-1, 1) means converting a one-dimensional array to a two-dimensional array with one column and multiple rows.
```

```python
x = np.array([
    [1, 10000000],
    [2, 20000000],
    [3, 30000000],
    [4, 40000000]
])

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)

# Fit means find the min and max values
# Another way to fit: scaler.fit(x)
# Transform means scale the data between 0 and 1 
# Another way to transform: x = scaler.transform(x)
# If you want to change the scale this parameter:: scaler = MinMaxScaler(feature_range=(min, max)) 
```

```python
# ===================================================================================
# standard scaling :
# ===================================================================================

# z = (x - mean(x)) / std(x)
# Standard scaler scales the data to have mean 0 and standard deviation 1.
# For Standard scaling, we (must) have a (two)-dimensional array.
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)

# Fit means find the mean and std values
# Another way to fit: scaler.fit(x)
# Transform means scale the data to have mean 0 and std 1
# Another way to transform: x = scaler.transform(x)
```

```python
# ===================================================================================
# Z-score scaling :
# ===================================================================================

# z = (x - mean(x)) / std(x)
# Z-score scaling is similar to standard scaling.
# The only difference is that Z-score scaling uses population standard deviation instead of sample standard deviation.
# However, in practice, both standard scaling and Z-score scaling are often used interchangeably.
# So, there is no separate implementation for Z-score scaling in sklearn.
# Therefore, you can use StandardScaler for Z-score scaling as well.
# But if you want to implement Z-score scaling with sample standard deviation, you can use the following code:

from scipy.stats import zscore
x = zscore(x, ddof=1)  # ddof=1 for sample standard deviation
```
